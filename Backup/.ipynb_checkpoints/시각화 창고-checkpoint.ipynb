```python
fig, ax = plt.subplots(figsize = (10, 10))
lines = []
colors = []

for veh in [1392, 1393]:
    sample = df[df['Vehicle ID'] == veh]

    x = np.array(sample['local_x'])
    y = np.array(sample['local_y'])
    cols = np.array(sample['ACTIT'])

    points = np.array([x, y]).T.reshape(-1, 1, 2)
    segments = np.concatenate([points[:-1], points[1:]], axis=1)

    lines.append(segments)
    colors.extend(cols)

all_lines = LineCollection(np.concatenate(lines), cmap='viridis')

# Calculate the minimum and maximum values of colors
min_val = np.min(colors)
max_val = np.max(colors)

# Normalize the color values based on the overall minimum and maximum
norm = plt.Normalize(min_val, max_val)

# Set the color array and normalization for all_lines
all_lines.set_array(np.array(colors))
all_lines.set_norm(norm)

ax.add_collection(all_lines)

ax.set_xlim(0, x.max())
ax.set_ylim(0, 50)

ax.axes.set_aspect('equal') # 1:1 비율로 같은 비율의 축척 사용

# Set title and labels
ax.set_title('Example plot')
ax.set_xlabel('x')
ax.set_ylabel('y')


plt.colorbar(all_lines, ax = ax, location = 'bottom')

plt.show()
```

```python

def graph_2d_reds(Vehicle_ID, LV_ID, SSM):
    """
    0에서 증가할수록 위험
    ['TIT', 'TIT2', 'ACTIT', 'TIDRAC', 'TIDSS', 'TERCRI', 'SSCR']
    """

    sample = df[(df['Vehicle ID'] == Vehicle_ID) & (df['LV_ID'] == LV_ID)]
    sample = sample.reset_index(drop = True)
    sample['Time'] = (sample['Time'] - sample['Time'].min()).round(2)

    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot()

    ## 대상차량 궤적 표시
    V_x = np.array(sample['local_x'])
    V_y = np.array(sample['local_y'])
    V_time = np.array(sample['Time'])
    V_ssm = np.array(sample[SSM])

    ## 선행차량 궤적 표시
    LV_x = np.array(sample['LV_local_x'])
    LV_y = np.array(sample['LV_local_y'])
    LV_time = np.array(sample['Time'])
    
    # 백그라운드 점 그래프 그리기
    LV_sdw = ax.scatter(LV_x, LV_y, c = 'lightgray', marker = '^')
    V_sdw = ax.scatter(V_x, V_y, c = 'lightgray', marker = 'o')

    # 점 그래프 그리기
    LV_trj= ax.scatter(LV_x, LV_y, c = V_ssm, marker = '^', edgecolor = 'gray', linewidth = 0.5, cmap = 'Reds', label = 'Leading Vehicle',
                       vmin = 0, vmax = np.nanmax(V_ssm))
    V_trj = ax.scatter(V_x, V_y, c = V_ssm, marker = 'o', edgecolor = 'gray', linewidth = 0.5, cmap = 'Reds', label = 'Following Vehicle',
                       vmin = 0, vmax = np.nanmax(V_ssm))


    # Annotation

    ## Following Vehicle Annotation
    time_max = math.floor(V_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={V_time[i*10]}', xy = (V_x[i*10], V_y[i*10]), xytext = (V_x[i*10] + 2, V_y[i*10] + 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ## Leading Vehicle Annotation
    time_max = math.floor(LV_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={LV_time[i*10]}', xy = (LV_x[i*10], LV_y[i*10]), xytext = (LV_x[i*10] - 2, LV_y[i*10] - 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ### 마지막 점 표시
    # ax.annotate(f't={LV_time[-1]}', xy = (LV_x[-1], LV_y[-1]), xytext = (LV_x[-1] - 3, LV_y[-1] - 1.5), 
    #             arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))




    # 축 및 범례 설정

    ax.legend()
    ax.set_xlim(0, 250)
    ax.set_ylim(7.5, 20)
    ax.set_xlabel('local_x')
    ax.set_ylabel('local_y')

    fig.colorbar(V_trj, ax = ax) # 범례


    ## Super Title, Ax Title 설정
    fig.suptitle(f'Vehicle Pair : ({Vehicle_ID}, {LV_ID}), SSM : {SSM}') # 총 지도
    #ax.set_title(f'{SSM} : threshold 0, 값이 클수록 위험함')
    
    plt.show()


def graph_2d_TTC(Vehicle_ID, LV_ID, SSM, threshold):
    """
    최소값 : 0, 임계값 : threshold, 최대값 : 최대
    임계값 이하, 0에 가까울수록 작은 경우
    ['TTC', 'T2', 'MTTC', 'pPET', 'TA', 'ACT', 'PSD']
    """

    sample = df[(df['Vehicle ID'] == Vehicle_ID) & (df['LV_ID'] == LV_ID)]
    sample = sample.reset_index(drop = True)
    sample['Time'] = (sample['Time'] - sample['Time'].min()).round(2)

    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot()

    ## 대상차량 궤적 표시
    V_x = np.array(sample['local_x'])
    V_y = np.array(sample['local_y'])
    V_time = np.array(sample['Time'])
    V_ssm = np.array(sample[SSM])

    ## 선행차량 궤적 표시
    LV_x = np.array(sample['LV_local_x'])
    LV_y = np.array(sample['LV_local_y'])
    LV_time = np.array(sample['Time'])


    # 백그라운드 점 그래프 그리기
    LV_sdw= ax.scatter(LV_x, LV_y, c = 'lightgray', marker = '^')
    V_sdw = ax.scatter(V_x, V_y, c = 'lightgray', marker = 'o')

    # 점 그래프 그리기

    divnorm = colors.TwoSlopeNorm(vmin = 0, vcenter = threshold, vmax = 10)

    LV_trj= ax.scatter(LV_x, LV_y, c = V_ssm, 
                       marker = '^', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'RdBu', norm = divnorm,
                       label = 'Leading Vehicle')
    V_trj = ax.scatter(V_x, V_y, c = V_ssm, 
                       marker = 'o', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'RdBu', norm = divnorm,
                       label = 'Following Vehicle')


    # Annotation

    ## Following Vehicle Annotation
    time_max = math.floor(V_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={V_time[i*10]}', xy = (V_x[i*10], V_y[i*10]), xytext = (V_x[i*10] + 2, V_y[i*10] + 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ## Leading Vehicle Annotation
    time_max = math.floor(LV_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={LV_time[i*10]}', xy = (LV_x[i*10], LV_y[i*10]), xytext = (LV_x[i*10] - 2, LV_y[i*10] - 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ### 마지막 점 표시
    # ax.annotate(f't={LV_time[-1]}', xy = (LV_x[-1], LV_y[-1]), xytext = (LV_x[-1] - 3, LV_y[-1] - 1.5), 
    #             arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))


    # 축 및 범례 설정

    ax.legend()
    ax.set_xlim(0, 250)
    ax.set_ylim(7.5, 20)
    ax.set_xlabel('local_x')
    ax.set_ylabel('local_y')

    # 컬러바 설정
    fig.colorbar(V_trj, ax = ax) # 범례


    ## Super Title, Ax Title 설정
    fig.suptitle(f'Vehicle Pair : ({Vehicle_ID}, {LV_ID}), SSM : {SSM}') # 총 지도
    #ax.set_title(f'{SSM} : threshold 0, 값이 클수록 위험함')

    plt.show()



def graph_2d_DSS(Vehicle_ID, LV_ID, SSM, threshold):
    """
    최소값 : 없음, 임계값 : threshold, 최대값 : 최대
    임계값 이하, 값이 낮을수록 위험한 경우
    임계값 위는 아무 문제없음(안전)
    ['PICUD', 'DSS']
    """
    
    SSM = 'DSS' # 임계값 아래, 값이 낮을수록 위험. 임계값 위는 아무 문제 없음(안전) 

    Vehicle_ID = 999
    LV_ID = 998
    threshold = 0

    sample = df[(df['Vehicle ID'] == Vehicle_ID) & (df['LV_ID'] == LV_ID)]
    sample = sample.reset_index(drop = True)
    sample['Time'] = (sample['Time'] - sample['Time'].min()).round(2)

    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot()

    ## 대상차량 궤적 표시
    V_x = np.array(sample['local_x'])
    V_y = np.array(sample['local_y'])
    V_time = np.array(sample['Time'])
    V_ssm = np.array(sample[SSM])

    ## 선행차량 궤적 표시
    LV_x = np.array(sample['LV_local_x'])
    LV_y = np.array(sample['LV_local_y'])
    LV_time = np.array(sample['Time'])


    # 백그라운드 점 그래프 그리기
    LV_sdw= ax.scatter(LV_x, LV_y, c = 'lightgray', marker = '^')
    V_sdw = ax.scatter(V_x, V_y, c = 'lightgray', marker = 'o')

    # 점 그래프 그리기

    divnorm = colors.TwoSlopeNorm(vmin = np.nanmin(V_ssm), vcenter = threshold, vmax = -1 * np.nanmin(V_ssm))

    LV_trj= ax.scatter(LV_x, LV_y, c = V_ssm, 
                       marker = '^', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'RdBu', norm = divnorm,
                       label = 'Leading Vehicle')
    V_trj = ax.scatter(V_x, V_y, c = V_ssm, 
                       marker = 'o', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'RdBu', norm = divnorm,
                       label = 'Following Vehicle')


    # Annotation

    ## Following Vehicle Annotation
    time_max = math.floor(V_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={V_time[i*10]}', xy = (V_x[i*10], V_y[i*10]), xytext = (V_x[i*10] + 2, V_y[i*10] + 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ## Leading Vehicle Annotation
    time_max = math.floor(LV_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={LV_time[i*10]}', xy = (LV_x[i*10], LV_y[i*10]), xytext = (LV_x[i*10] - 2, LV_y[i*10] - 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ### 마지막 점 표시
    # ax.annotate(f't={LV_time[-1]}', xy = (LV_x[-1], LV_y[-1]), xytext = (LV_x[-1] - 3, LV_y[-1] - 1.5), 
    #             arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))


    # 축 및 범례 설정

    ax.legend()
    ax.set_xlim(0, 250)
    ax.set_ylim(7.5, 20)
    ax.set_xlabel('local_x')
    ax.set_ylabel('local_y')

    # 컬러바 설정
    fig.colorbar(V_trj, ax = ax) # 범례


    ## Super Title, Ax Title 설정
    fig.suptitle(f'Vehicle Pair : ({Vehicle_ID}, {LV_ID}), SSM : {SSM}') # 총 지도
    #ax.set_title(f'{SSM} : threshold 0, 값이 클수록 위험함')

    plt.show()



def graph_2d_DRAC(Vehicle_ID, LV_ID, SSM, threshold):
    """
    임계값 위, 값이 높을수록 위험. 임계값 아래는 아무 문제 없음
    ['DRAC', 'MDRAC', 'SSCR', 'RCRI']
    """
    
    SSM = 'DRAC' # 임계값 위, 값이 높을수록 위험. 임계값 아래는 아무 문제 없음(안전) 

    Vehicle_ID = 999
    LV_ID = 998
    threshold = 3.4

    sample = df[(df['Vehicle ID'] == Vehicle_ID) & (df['LV_ID'] == LV_ID)]
    sample = sample.reset_index(drop = True)
    sample['Time'] = (sample['Time'] - sample['Time'].min()).round(2)

    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot()

    ## 대상차량 궤적 표시
    V_x = np.array(sample['local_x'])
    V_y = np.array(sample['local_y'])
    V_time = np.array(sample['Time'])
    V_ssm = np.array(sample[SSM])

    ## 선행차량 궤적 표시
    LV_x = np.array(sample['LV_local_x'])
    LV_y = np.array(sample['LV_local_y'])
    LV_time = np.array(sample['Time'])


    # 백그라운드 점 그래프 그리기
    LV_sdw= ax.scatter(LV_x, LV_y, c = 'lightgray', marker = '^')
    V_sdw = ax.scatter(V_x, V_y, c = 'lightgray', marker = 'o')

    # 점 그래프 그리기

    divnorm = colors.TwoSlopeNorm(vmin = np.nanmin(V_ssm), vcenter = threshold, vmax = threshold + 5)

    LV_trj= ax.scatter(LV_x, LV_y, c = V_ssm, 
                       marker = '^', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'turbo', norm = divnorm,
                       label = 'Leading Vehicle')
    V_trj = ax.scatter(V_x, V_y, c = V_ssm, 
                       marker = 'o', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'turbo', norm = divnorm,
                       label = 'Following Vehicle')


    # Annotation

    ## Following Vehicle Annotation
    time_max = math.floor(V_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={V_time[i*10]}', xy = (V_x[i*10], V_y[i*10]), xytext = (V_x[i*10] + 2, V_y[i*10] + 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ## Leading Vehicle Annotation
    time_max = math.floor(LV_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={LV_time[i*10]}', xy = (LV_x[i*10], LV_y[i*10]), xytext = (LV_x[i*10] - 2, LV_y[i*10] - 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ### 마지막 점 표시
    # ax.annotate(f't={LV_time[-1]}', xy = (LV_x[-1], LV_y[-1]), xytext = (LV_x[-1] - 3, LV_y[-1] - 1.5), 
    #             arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))


    # 축 및 범례 설정

    ax.legend()
    ax.set_xlim(0, 250)
    ax.set_ylim(7.5, 20)
    ax.set_xlabel('local_x')
    ax.set_ylabel('local_y')

    # 컬러바 설정
    fig.colorbar(V_trj, ax = ax) # 범례


    ## Super Title, Ax Title 설정
    fig.suptitle(f'Vehicle Pair : ({Vehicle_ID}, {LV_ID}), SSM : {SSM}') # 총 지도
    #ax.set_title(f'{SSM} : threshold 0, 값이 클수록 위험함')

    plt.show()


def graph_2d_other(Vehicle_ID, LV_ID, SSM):
    """
    임계값이 분명하지 않은 경우, turbo로 분포만 확인함
    ['TA_CS', 'CS', 'MTC', 'MMTC', 'DCIA', 'Unsafety']
    
    """
    sample = df[(df['Vehicle ID'] == Vehicle_ID) & (df['LV_ID'] == LV_ID)]
    sample = sample.reset_index(drop = True)
    sample['Time'] = (sample['Time'] - sample['Time'].min()).round(2)

    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot()

    ## 대상차량 궤적 표시
    V_x = np.array(sample['local_x'])
    V_y = np.array(sample['local_y'])
    V_time = np.array(sample['Time'])
    V_ssm = np.array(sample[SSM])

    ## 선행차량 궤적 표시
    LV_x = np.array(sample['LV_local_x'])
    LV_y = np.array(sample['LV_local_y'])
    LV_time = np.array(sample['Time'])


    # 백그라운드 점 그래프 그리기
    LV_sdw= ax.scatter(LV_x, LV_y, c = 'lightgray', marker = '^')
    V_sdw = ax.scatter(V_x, V_y, c = 'lightgray', marker = 'o')

    # 점 그래프 그리기

    LV_trj= ax.scatter(LV_x, LV_y, c = V_ssm, 
                       marker = '^', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'turbo', vmin = np.nanmin(V_ssm), vmax = np.nanmax(V_ssm),
                       label = 'Leading Vehicle')
    V_trj = ax.scatter(V_x, V_y, c = V_ssm, 
                       marker = 'o', edgecolor = 'gray', linewidth = 0.5, 
                       cmap = 'turbo', vmin = np.nanmin(V_ssm), vmax = np.nanmax(V_ssm),
                       label = 'Following Vehicle')


    # Annotation

    ## Following Vehicle Annotation
    time_max = math.floor(V_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={V_time[i*10]}', xy = (V_x[i*10], V_y[i*10]), xytext = (V_x[i*10] + 2, V_y[i*10] + 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ## Leading Vehicle Annotation
    time_max = math.floor(LV_time[-1])

    ### 1초 단위로 점 표시
    for i in range(time_max+1):
        ax.annotate(f't={LV_time[i*10]}', xy = (LV_x[i*10], LV_y[i*10]), xytext = (LV_x[i*10] - 2, LV_y[i*10] - 1), 
                    arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))

    ### 마지막 점 표시
    # ax.annotate(f't={LV_time[-1]}', xy = (LV_x[-1], LV_y[-1]), xytext = (LV_x[-1] - 3, LV_y[-1] - 1.5), 
    #             arrowprops = dict(facecolor = 'black', arrowstyle = '-|>'))


    # 축 및 범례 설정

    ax.legend()
    ax.set_xlim(0, 250)
    ax.set_ylim(7.5, 20)
    ax.set_xlabel('local_x')
    ax.set_ylabel('local_y')

    # 컬러바 설정
    fig.colorbar(V_trj, ax = ax) # 범례


    ## Super Title, Ax Title 설정
    fig.suptitle(f'Vehicle Pair : ({Vehicle_ID}, {LV_ID}), SSM : {SSM}') # 총 지도
    #ax.set_title(f'{SSM} : threshold 0, 값이 클수록 위험함')

    plt.show()
```

```python
class Annotation3D(Annotation):

    def __init__(self, text, xyz, *args, **kwargs):
        super().__init__(text, xy=(0, 0), *args, **kwargs)
        self._xyz = xyz

    def draw(self, renderer):
        x2, y2, z2 = proj_transform(*self._xyz, self.axes.M)
        self.xy = (x2, y2)
        super().draw(renderer)
        
def _annotate3D(ax, text, xyz, *args, **kwargs):
    '''Add anotation `text` to an `Axes3d` instance.'''

    annotation = Annotation3D(text, xyz, *args, **kwargs)
    ax.add_artist(annotation)

setattr(Axes3D, 'annotate3D', _annotate3D)
```
